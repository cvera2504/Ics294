5. (15 puntos) Usando la regresión del salario en función de sales, roe y ros, se distinguen los
grupos mediante las dummies; indus, finance, consprod y utility, que corresponden al tipo de
empresa. Realice mediante test de Chow, con 5% de significancia, si los modelos para cada
grupo son iguales.

```r
# Variables independientes en el modelo base (excluyendo el intercepto)
k <- 3

# Crear subconjuntos de datos para cada tipo de industria
df_indus <- subset(df_ceosal1, indus == 1)
df_finance <- subset(df_ceosal1, finance == 1)
df_consprod <- subset(df_ceosal1, consprod == 1)
df_utility <- subset(df_ceosal1, utility == 1)

# Lista para almacenar los SSR individuales y los modelos
ssr_individuales <- c()
modelos_individuales <- list()

# Número de grupos con suficientes observaciones
num_grupos_validos <- 0

# Procesar grupo 'indus'
if (nrow(df_indus) > k + 1) {
    model_indus <- lm(lsalary ~ sales + roe + ros, data = df_indus)
    ssr_individuales["indus"] <- sum(residuals(model_indus)^2)
    modelos_individuales[["indus"]] <- model_indus
    num_grupos_validos <- num_grupos_validos + 1
} else {
    print("Advertencia: El grupo 'indus' tiene muy pocas observaciones para una regresión válida.")
}

# Procesar grupo 'finance'
if (nrow(df_finance) > k + 1) {
    model_finance <- lm(lsalary ~ sales + roe + ros, data = df_finance)
    ssr_individuales["finance"] <- sum(residuals(model_finance)^2)
    modelos_individuales[["finance"]] <- model_finance
    num_grupos_validos <- num_grupos_validos + 1
} else {
    print("Advertencia: El grupo 'finance' tiene muy pocas observaciones para una regresión válida.")
}

# Procesar grupo 'consprod'
if (nrow(df_consprod) > k + 1) {
    model_consprod <- lm(lsalary ~ sales + roe + ros, data = df_consprod)
    ssr_individuales["consprod"] <- sum(residuals(model_consprod)^2)
    modelos_individuales[["consprod"]] <- model_consprod
    num_grupos_validos <- num_grupos_validos + 1
} else {
    print("Advertencia: El grupo 'consprod' tiene muy pocas observaciones para una regresión válida.")
}

# Procesar grupo 'utility'
if (nrow(df_utility) > k + 1) {
    model_utility <- lm(lsalary ~ sales + roe + ros, data = df_utility)
    ssr_individuales["utility"] <- sum(residuals(model_utility)^2)
    modelos_individuales[["utility"]] <- model_utility
    num_grupos_validos <- num_grupos_validos + 1
} else {
    print("Advertencia: El grupo 'utility' tiene muy pocas observaciones para una regresión válida.")
}

# Verificar si hay al menos dos grupos válidos para realizar el test de Chow
if (num_grupos_validos < 2) {
    stop("No hay suficientes grupos válidos para realizar el Test de Chow.")
}

# Suma de cuadrados de los residuos del modelo no restringido (suma de los SSR de cada grupo)
SSR_unrestricted <- sum(ssr_individuales)

# Suma de cuadrados de los residuos del modelo restringido (modelo original sin diferencias de grupo)
# model_salary ya está definido como lsalary ~ sales + roe + ros
SSR_restricted <- sum(residuals(model_salary)^2)

# Número total de observaciones
n <- nrow(df_ceosal1)

# Número de parámetros en cada modelo de grupo (k variables + 1 intercepto)
p <- k + 1

# Grados de libertad del numerador: (G - 1) * p
gl1 <- (num_grupos_validos - 1) * p

# Grados de libertad del denominador: n - G * p
gl2 <- n - (num_grupos_validos * p)

# Calcular el estadístico F del Chow test
F_statistic_chow <- ((SSR_restricted - SSR_unrestricted) / gl1) / (SSR_unrestricted / gl2)

# Calcular el p-valor
p_value_chow <- 1 - pf(F_statistic_chow, gl1, gl2)

# Nivel de significancia
alfa <- 0.05

cat("\n--- Resultados del Test de Chow ---\n")
cat(paste("SSR Restringido (modelo completo): ", round(SSR_restricted, 4), "\n"))
cat(paste("SSR No Restringido (suma de modelos por grupo): ", round(SSR_unrestricted, 4), "\n"))
cat(paste("Estadístico F del Chow test: ", round(F_statistic_chow, 4), "\n"))
cat(paste("Grados de libertad del numerador (gl1): ", gl1, "\n"))
cat(paste("Grados de libertad del denominador (gl2): ", gl2, "\n"))
cat(paste("P-valor del Test de Chow: ", round(p_value_chow, 4), "\n"))

if (p_value_chow < alfa) {
    cat(paste("Dado que el p-valor (", round(p_value_chow, 4), ") es menor que el nivel de significancia (", alfa, "), rechazamos la hipótesis nula.", "\n"))
    cat("Esto significa que los modelos de regresión son significativamente diferentes entre los grupos de la industria.\n")
} else {
    cat(paste("Dado que el p-valor (", round(p_value_chow, 4), ") es mayor o igual que el nivel de significancia (", alfa, "), no rechazamos la hipótesis nula.", "\n"))
    cat("Esto significa que no hay evidencia suficiente para concluir que los modelos de regresión son significativamente diferentes entre los grupos de la industria.\n")
}

# Opcional: Imprimir resúmenes de los modelos individuales si se desea revisar
# for (name in names(modelos_individuales)) {
#   print(paste("\n--- Resumen del Modelo para el grupo", name, "---"))
#   print(summary(modelos_individuales[[name]]))
# }
```
